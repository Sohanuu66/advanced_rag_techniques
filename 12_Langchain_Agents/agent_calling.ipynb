```python
import os
from dotenv import load_dotenv

from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
api_key = os.getenv('GEMINI_API_KEY')
model = ChatGoogleGenerativeAI(model='gemini-2.5-flash-lite', api_key=api_key)
```

```python
import langchain
print(langchain.__version__)
```

```python
# AgentExecutor doesnt lie in this version of langchain as now these type of agents are made in langGraph
```

```python
from langchain_core.tools import tool

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model=model,
    tools=[multiply],
)
# internally it uses langgraph and uses AgentExecutor also
# create_react_agent is depricated --> create_agent automatically handles it
```

```python
response = agent.invoke(
    {
        "messages": [
            {"role": "user", "content": "What is 4 times 6?"}
        ]
    }
)

print(response["messages"][-1].content)
```
